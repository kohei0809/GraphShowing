```python
# 回収したごみの量をグラフにするプログラム

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pathlib

#各種設定
agent_number_list = [8, 9, 10, 15]
#agent_number_list = [4, 11, 7, 18]
n = 0

date = "09-17 14-56-24 robots-number=20_1"
#targetType = "0-Random"
#targetType = "1-Greedy"
#targetType = "2-Interval"
#targetType = "3-BNPS"
#targetType = "4-AMTDS-LD"
targetType = "5-AMTDS-ESC"
agent_number = len(agent_number_list)
i = 0

df = pd.read_csv("../../../../Volumes/ExternalDisk1/Patrolling/log/OTest/Onebyone/Office/" + targetType + "/" + date + "/AccumulatedLitter.csv", index_col=0)

for number in agent_number_list:
    plt.plot(df.index, df[str(number)], color=cm.hsv(i/float(agent_number)), label="Agent"+str(number))
    i += 1
    
#ラベルの追加
plt.xlabel('Time(steps)')
plt.ylabel('Accumulated Litter Amount')

#表示範囲の指定
plt.xlim(0, 5000000)
plt.ylim(0, 500)

plt.xticks([0, 1000000, 2000000, 3000000, 4000000, 5000000])

#凡例の追加
plt.legend()

#指数表記から普通の表記に変換
plt.ticklabel_format(style='plain',axis='x')
plt.ticklabel_format(style='plain',axis='y')

#ファルダがない場合は作成
p_dir = pathlib.Path("../../../../Volumes/ExternalDisk1/Patrolling/result/al_graph/TTest/" + targetType)
if not p_dir.exists():
    p_dir.mkdir(parents=True)
    
#グラフの保存
plt.savefig(str(p_dir) + '/' + 'AccumulatedLitter' + date + '_' + str(n) + '.png')

#グラフの表示
plt.show()

print("Showing Accumulated Litter graph is completed.")
```
