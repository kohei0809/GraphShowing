```python
#StrategyDistribution(間隔33600ステップでの核戦略を取ったエージェント数の記録)のcsvファイルからエージェント数の時間変化の平均値のグラフを出力するプログラム

import pandas as pd
import matplotlib.pyplot as plt
import pathlib

#設定
file_count = 25
date = "07-03 20-52-05 robots-number=20"

for i in range(1, file_count + 1):
    df = pd.read_csv("../../../../Volumes/ExternalDisk1/Patrolling/log/TTest/TimeChange/Office/5-AMTDS-ESC/" + date + "_" + str(i) + "/StrategyDistribution.csv")
    if(i == 1):
        df_sum = df
    else:
        df_sum += df
        
#平均値計算
ave_df = df_sum / file_count

#プロット
plt.plot(ave_df['time'], ave_df['R'], color="black", label="Random")
plt.plot(ave_df['time'], ave_df['PGS'], color="red", label="Greedy")
plt.plot(ave_df['time'], ave_df['LI'], color="green", label="Interval")
plt.plot(ave_df['time'], ave_df['BNPS'], color="blue", label="BNPS")

#ラベルの追加
plt.xlabel('Time(ticks)')
plt.ylabel('Agent Number')

#表示範囲の指定
plt.xlim(0, 3500000)
plt.ylim(0, 20)

#凡例の追加
plt.legend()

#指数表記から普通の表記に変換
plt.ticklabel_format(style='plain',axis='x')
plt.ticklabel_format(style='plain',axis='y')

#フォルダがない場合は、作成
p_dir = pathlib.Path("../../../../Volumes/ExternalDisk1/Patrolling/result/strategy_ave")
if not p_dir.exists():
    p_dir.mkdir(parents=True)

#グラフの保存
plt.savefig('../../../../Volumes/ExternalDisk1/Patrolling/result/strategy_ave/' + 'strategy_graph_3600_ave_' + str(file_count) + "_" + date + ".png")

#グラフの表示
plt.show()

print("Showing Strategy Average graph is completed.")
```
