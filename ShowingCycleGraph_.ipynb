```python
#cycle(各サイクルを選択したエージェント数の記録)のcsvファイルからエージェント数の時間変化の平均値のグラフを出力するプログラム

import pandas as pd
import matplotlib.pyplot as plt
import pathlib

#設定
file_count = 10
date = "09-11 01-27-53 robots-number=20"

for i in range(1, file_count + 1):
    df = pd.read_csv("../../../../Volumes/ExternalDisk1/Patrolling/log/UTest/CycleLearning/Office/5-AMTDS-ESC/" + date + "_" + str(i) + "/Cycle.csv", names=['time','150', '300', '900'] , header=None)
    if(i == 1):
        df_sum = df
    else:
        df_sum += df
        
#print(df_sum)
#平均値計算
ave_df = df_sum / file_count

#プロット
plt.plot(ave_df['time'], ave_df['150'], color="black", label="150")
plt.plot(ave_df['time'], ave_df['300'], color="red", label="300")
plt.plot(ave_df['time'], ave_df['900'], color="blue", label="900")

#ラベルの追加
plt.xlabel('Time(ticks)')
plt.ylabel('Agent Number')

#表示範囲の指定
#plt.xlim(0, 2000000)
plt.ylim(0, 20)

#凡例の追加
plt.legend()

#指数表記から普通の表記に変換
plt.ticklabel_format(style='plain',axis='x')
plt.ticklabel_format(style='plain',axis='y')

#フォルダがない場合は、作成
p_dir = pathlib.Path("../../../../Volumes/ExternalDisk1/Patrolling/result/Cycle")
if not p_dir.exists():
    p_dir.mkdir(parents=True)

#グラフの保存
plt.savefig('../../../../Volumes/ExternalDisk1/Patrolling/result/Cycle/' + 'cycle_graph_' + str(file_count) + "_" + date + ".png")

#グラフの表示
plt.show()

print("Showing Strategy Average graph is completed.")
```
