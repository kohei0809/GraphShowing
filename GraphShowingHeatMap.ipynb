```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#設定
file_count = 10
#env = "Block"
#env = "Complex"
env = "Office"

#date = "11-13 05-11-19"
#date = "11-14 04-08-49"
date = "11-14 09-50-22"

scale = 50
size = scale * 2 + 1


#データ読み込み
data = [0 for i in range(file_count)]
name = "log/ETest/" + env + "/" + date + "_"

for i in range(file_count):
    data[i] = np.loadtxt(name + str(i+1) + "/LitterAmount/LitterAmount_1500000.csv", dtype="int", delimiter= ",")

#平均値出力
data_ave = [[0 for i in range(size)] for j in range(size)]

for i in range(size):
    for j in range(size):
        sum = 0
        for k in range(file_count): 
            sum = sum + data[k][i][j]
        data_ave[i][j] = sum / file_count

sns.heatmap(data_ave, cmap="Reds", center=1300, xticklabels=False, yticklabels=False)

#画像を保存
name = "log/ETest/" + env + "/Graph_" + env
plt.savefig(name + "_heat150000.png")
print("Graph showing completed.")
```

```python
from PIL import Image, ImageDraw
import numpy as np
import pathlib
import matplotlib.pyplot as plt

#各種設定
agent_number = 20
date = "12-10 00-01-31"  

time = 0
scale = 2
px = 0
py = 0
px = px + scale
py = -py + scale

filename = "./log/test/Test/" + date + "/LitterAmount/LitterAmount_" + str(time) + ".csv"        
        
#filenameのデータを配列に格納
data = np.loadtxt(filename, dtype="int", delimiter= ",")
print(data)
        
#配列dataのサイズの画像を作成
height, width = data.shape
img = Image.new("RGB", (width*20, height*20))
        
#格納されている数字に従って画像の色を設定
for y in range(height):
    for x in range(width):
        number = data[y][x]
        #ごみの量が多いほど色を濃くする
        if number == 0:
            for z in range(0, 20):
                for w in range(0, 20):
                    if (z < 1) or (w < 1):
                        img.putpixel((x*20+z, y*20+w), (0, 0, 0))
                    else:
                        img.putpixel((x*20+z, y*20+w), (255, 255, 255))
        elif number == 1:
             for z in range(0, 20):
                for w in range(0, 20):
                    if (z < 1) or (w < 1):
                        img.putpixel((x*20+z, y*20+w), (0, 0, 0))
                    else:
                        img.putpixel((x*20+z, y*20+w), (255, 229, 229))
        elif number == 2:
             for z in range(0, 20):
                for w in range(0, 20):
                    if (z < 1) or (w < 1):
                        img.putpixel((x*20+z, y*20+w), (0, 0, 0))
                    else:
                        img.putpixel((x*20+z, y*20+w), (255, 204, 204))
        elif number == 3:
             for z in range(0, 20):
                for w in range(0, 20):
                    if (z < 1) or (w < 1):
                        img.putpixel((x*20+z, y*20+w), (0, 0, 0))
                    else:
                        img.putpixel((x*20+z, y*20+w), (255, 178, 178))
        elif number == 4:
             for z in range(0, 20):
                for w in range(0, 20):
                    if (z < 1) or (w < 1):
                        img.putpixel((x*20+z, y*20+w), (0, 0, 0))
                    else:
                        img.putpixel((x*20+z, y*20+w), (255, 153, 153))
        elif number == 5:
             for z in range(0, 20):
                for w in range(0, 20):
                    if (z < 1) or (w < 1):
                        img.putpixel((x*20+z, y*20+w), (0, 0, 0))
                    else:
                        img.putpixel((x*20+z, y*20+w), (255, 127, 127))
        elif number == 6:
             for z in range(0, 20):
                for w in range(0, 20):
                    if (z < 1) or (w < 1):
                        img.putpixel((x*20+z, y*20+w), (0, 0, 0))
                    else:
                        img.putpixel((x*20+z, y*20+w), (255, 102, 102))
        elif number == 7:
             for z in range(0, 20):
                for w in range(0, 20):
                    if (z < 1) or (w < 1):
                        img.putpixel((x*20+z, y*20+w), (0, 0, 0))
                    else:
                        img.putpixel((x*20+z, y*20+w), (255, 76, 76))
        elif number == 8:
             for z in range(0, 20):
                for w in range(0, 20):
                    if (z < 1) or (w < 1):
                        img.putpixel((x*20+z, y*20+w), (0, 0, 0))
                    else:
                        img.putpixel((x*20+z, y*20+w), (255, 50, 50))
        elif number == 9:
             for z in range(0, 20):
                for w in range(0, 20):
                    if (z < 1) or (w < 1):
                        img.putpixel((x*20+z, y*20+w), (0, 0, 0))
                    else:
                        img.putpixel((x*20+z, y*20+w), (255, 25, 25))
        elif number == 10:
             for z in range(0, 20):
                for w in range(0, 20):
                    if (z < 1) or (w < 1):
                        img.putpixel((x*20+z, y*20+w), (0, 0, 0))
                    else:
                        img.putpixel((x*20+z, y*20+w), (255, 0, 0))
        elif number > 10:
             for z in range(0, 20):
                for w in range(0, 20):
                    if (z < 1) or (w < 1):
                        img.putpixel((x*20+z, y*20+w), (0, 0, 0))
                    else:
                        img.putpixel((x*20+z, y*20+w), (178, 0, 0))
        else:
            print("({0},{1})でエラー".format(x,y))
            
#自己位置を青色でマーク
for z in range(0, 20):
    for w in range(0, 20):
        if (z < 1) or (w < 1):
            img.putpixel((px*20+z, py*20+w), (0, 0, 0))
        else:
            img.putpixel((px*20+z, py*20+w), (0, 0, 255))
            
#外枠の残りを黒くする
for y in range(height):
    for x in range(width):
        if(x == width-1) or (y == height-1):
            for z in range(0, 20):
                    for w in range(0, 20):
                        if (x == width-1 and z > 18) or (y == height-1 and w > 18):
                            img.putpixel((x*20+z, y*20+w), (0, 0, 0))
    
#フォルダがない場合は、作成 
p_dir = pathlib.Path("./result/ht_map")
if not p_dir.exists():
    p_dir.mkdir(parents=True)
#画像を表示
img.save("./result/ht_map/heatmap_" + str(time) + ".png")

#svgファイルの出力

# 画像をNumpy配列に変換する
image_narray = np.asarray(img)
    
#座標を削除
plt.axis('off')
    
#画像を表示
plt.imshow(image_narray)
               
#画像を保存
plt.savefig("./result/ht_map/heatmap_" + str(time) + ".eps")
print("Graph showing completed.")
```
