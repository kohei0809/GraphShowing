```python
#evaluation-s3600(間隔3600ステップでのD(s)計算)のcsvファイルからD(s)の時間変化のグラフを出力するプログラム
import pandas as pd
import matplotlib.pyplot as plt
import pathlib

Request_flag = True

agent_type = "Onebyone"
target_type = "5-AMTDS-ESC"
date = "07-18 14-15-30 robots-number=20"

file_count = 4

df = pd.read_csv("../../../../Volumes/ExternalDisk1/Patrolling/log/MTest/" + agent_type + "/Office/" + target_type + "/" + date + "_" + str(file_count) + "/evaluation-s3600.csv", names=['time', 'ds', 'maxLitteerAmount', 'cs'], header=None)
plt.plot(df['time'], df['ds'], color="red", label="LD")

if Request_flag == True:
    request = 600.0
    #request = 1000.0
    df_req = pd.DataFrame([[0.0, request*3600, 0.0, 0.0], [5000000.0, request*3600, 0.0, 0.0]], columns=['time', 'ds', 'maxLitteerAmount', 'energy'])
    plt.plot(df_req['time'], df_req['ds'], color="green", label="Request")


#ラベルの追加
plt.xlabel('Time(ticks)')
plt.ylabel('Cumulative existance duration of events D(s)')

#表示範囲の指定
plt.xlim(0, 5000000)
#plt.ylim(0, 4500000)

#凡例の追加
#plt.legend()

#指数表記から普通の表記に変換
plt.ticklabel_format(style='plain',axis='x')
plt.ticklabel_format(style='plain',axis='y')

#フォルダがない場合は、作成
p_dir = pathlib.Path("./result/ds_graph/test")
if not p_dir.exists():
    p_dir.mkdir(parents=True)

#グラフの保存
plt.savefig('./result/ds_graph/test/' + 'ds_graph_3600_' + date + '.jpg')

#グラフの表示
plt.show()

print("Showing D(s) graph is completed.")
```
