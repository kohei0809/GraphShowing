```python
from PIL import Image, ImageDraw
import numpy as np
import pathlib
import matplotlib.pyplot as plt

#各種設定
agent_number = 20
agent_type = "Communicating"
target_type = "5-AMTDS-ESC"
date = "04-03 18-29-29" 

time = 0

for i in range(agent_number):
    filename = "./log/TTest/" + agent_type + "/Office/" + target_type + "/" + date + " robots-number=20_1/Agent" + str(i) + "/PositionCount/PositionCount_" + str(time) + ".csv"        
        
    #filenameのデータを配列に格納
    data = np.loadtxt(filename, dtype="int", delimiter= ",")
    #print(data)
        
    #配列dataのサイズの画像を作成
    height, width = data.shape
    img = Image.new("RGB", (width, height))
        
#格納されている数字に従って画像の色を設定
    for y in range(height):
        for x in range(width):
            number = data[y][x]
            
            if number == -1:
                img.putpixel((x, y),(0 , 0 , 0))
            elif number == 0:
                img.putpixel((x, y),(255, 255, 255))
            elif number > 0 and number < 25:
                img.putpixel((x, y),(255, 255, 200))
            elif number >= 25 and number < 50:
                img.putpixel((x, y),(255, 200, 100))
            elif number >= 50 and number < 75:
                img.putpixel((x, y),(255, 165, 50))
            elif number >= 75 and number < 100:
                img.putpixel((x, y),(255, 100, 0))
            elif number >= 100 and number < 150:
                img.putpixel((x, y),(255, 50, 0))
            elif number >= 150 and number < 200:
                img.putpixel((x, y),(255, 0, 0))
            elif number >= 200:
                img.putpixel((x, y), (178, 0, 0))
            else:
                print("({0},{1})でエラー".format(x,y))

    #print("{0},{1}".format(width*5, height*5))
    #そのままだと小さいので、5倍の大きさにする
    img = img.resize(((width*5), (height*5)))

    #フォルダがない場合は、作成
    p = pathlib.Path("./result/agent_position/" +agent_type + "/" + date + "/Agent" + str(i))
    if not p.exists():
        p.mkdir(parents=True)
    
    #指定したファイルに画像を保存
    new_filename = "graph" + "_Agent" + str(i) + ".png"
    img.save(str(p) + "/" + new_filename)

    # 画像をNumpy配列に変換する
    image_narray = np.asarray(img)
    
    #座標を削除
    plt.axis('off')
    
    #画像を表示
    plt.imshow(image_narray)


print("Graph showing completed.")
```

```python
from PIL import Image, ImageDraw
import numpy as np
import pathlib
import matplotlib.pyplot as plt

#各種設定
agent_number = 20
agent_type = "TimeChange_Learning"
target_type = "5-AMTDS-ESC"
date = "03-28 04-18-09" 

before_time = 3000000
after_time = 4000000

for i in range(agent_number):
    filename1 = "../../../../Volumes/ExternalDisk1/Patrolling/log/TTest/" + agent_type + "/Office/" + target_type + "/" + date + " robots-number=20_1/Agent" + str(i) + "/PositionCount/PositionCount_" + str(before_time) + ".csv"        
    filename2 = "../../../../Volumes/ExternalDisk1/Patrolling/log/TTest/" + agent_type + "/Office/" + target_type + "/" + date + " robots-number=20_1/Agent" + str(i) + "/PositionCount/PositionCount_" + str(after_time) + ".csv"

    #filenameのデータを配列に格納
    data1 = np.loadtxt(filename1, dtype="int", delimiter= ",")
    data2 = np.loadtxt(filename2, dtype="int", delimiter= ",")
    data = data2 - data1
        
    #配列dataのサイズの画像を作成
    height, width = data.shape
    img = Image.new("RGB", (width, height))
        
#格納されている数字に従って画像の色を設定
    for y in range(height):
        for x in range(width):
            number = data[y][x]
            exclude = data1[y][x]
            
            if exclude == -1:
                img.putpixel((x, y),(0 , 0 , 0))
            elif number == 0:
                img.putpixel((x, y),(255, 255, 255))
            elif number > 0 and number < 20:
                img.putpixel((x, y),(255, 220, 200))
            elif number >= 20 and number < 40:
                img.putpixel((x, y),(255, 165, 50))
            elif number >= 40 and number < 60:
                img.putpixel((x, y),(255, 100, 0))
            elif number >= 60 and number < 80:
                img.putpixel((x, y),(255, 50, 0))
            elif number >= 80 and number < 100:
                img.putpixel((x, y),(255, 0, 0))
            elif number >= 100:
                img.putpixel((x, y), (178, 0, 0))
            else:
                print("({0},{1})でエラー".format(x,y))

    #print("{0},{1}".format(width*5, height*5))
    #そのままだと小さいので、5倍の大きさにする
    img = img.resize(((width*5), (height*5)))

    #フォルダがない場合は、作成
    p = pathlib.Path("../../../../Volumes/ExternalDisk1/Patrolling/result/agent_position/" +agent_type + "/" + date + "/Agent" + str(i))
    if not p.exists():
        p.mkdir(parents=True)
    
    #指定したファイルに画像を保存
    new_filename = "graph" + "_Agent" + str(i) + ".png"
    img.save(str(p) + "/" + new_filename)

    # 画像をNumpy配列に変換する
    image_narray = np.asarray(img)
    
    #座標を削除
    plt.axis('off')
    
    #画像を表示
    plt.imshow(image_narray)


print("Graph showing completed.")
```
